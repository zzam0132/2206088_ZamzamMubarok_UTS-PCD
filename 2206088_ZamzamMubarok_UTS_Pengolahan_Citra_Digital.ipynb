```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Membaca citra grayscale
image_path = '/content/zam.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# 2. Menghitung histogram
histogram, bins = np.histogram(image.flatten(), bins=256, range=[0, 256])

# 3. Menampilkan histogram
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Histogram Citra Grayscale')
plt.xlabel('Intensitas Piksel')
plt.ylabel('Frekuensi')
plt.xlim([0, 256])
plt.plot(histogram, color='black')

# 4. Melakukan normalisasi histogram
normalized_histogram = histogram / histogram.sum()

# Menampilkan histogram yang dinormalisasi
plt.subplot(1, 2, 2)
plt.title('Histogram Normalisasi')
plt.xlabel('Intensitas Piksel')
plt.ylabel('Frekuensi Normalisasi')
plt.xlim([0, 256])
plt.plot(normalized_histogram, color='black')

# Menampilkan citra dan histogram
plt.tight_layout()
plt.show()
```

Penjelasan : Kode
*  Membaca Citra: Menggunakan cv2.imread() untuk membaca citra dalam mode grayscale.
*  Menghitung Histogram: np.histogram() digunakan untuk menghitung frekuensi setiap nilai intensitas piksel dari 0 hingga 255.

*  Menampilkan Histogram: Menggunakan Matplotli

*  b untuk membuat dua subplot; satu untuk histogram asli dan satu untuk histogram yang dinormalisasi.
*  Normalisasi Histogram: Histogram dinormalisasi dengan membagi setiap frekuensi dengan jumlah total piksel.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Membaca gambar
image_path = '/content/zam.jpg'
image = cv2.imread(image_path)

# 2. Mendefinisikan kernel rata-rata 3x3
kernel = np.ones((3, 3), np.float32) / 9

# 3. Melakukan operasi konvolusi
convolved_image = cv2.filter2D(image, -1, kernel)

# 4. Menampilkan gambar asli dan hasil konvolusi
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Gambar Asli')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Hasil Konvolusi (Kernel Rata-rata 3x3)')
plt.imshow(cv2.cvtColor(convolved_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.tight_layout()
plt.show()
```

Penjelasan : Kode
*  Membaca Gambar: Menggunakan cv2.imread() untuk membaca gambar dari path yang ditentukan.
*  Mendefinisikan Kernel: Kernel rata-rata 3x3 didefinisikan sebagai matriks dengan semua elemen bernilai 1/9, yang akan menghitung rata-rata dari piksel di sekitarnya.
*  Operasi Konvolusi: cv2.filter2D() digunakan untuk menerapkan kernel pada gambar, menghasilkan gambar yang telah diproses.
*  Menampilkan Gambar: Menggunakan Matplotlib untuk menampilkan gambar asli dan hasil konvolusi berdampingan.

Efek yang Terjadi Setelah Konvolusi
Setelah menerapkan konvolusi dengan kernel rata-rata 3x3 :
*  Penghalusan Gambar: Gambar menjadi lebih halus karena setiap piksel baru merupakan rata-rata dari piksel-piksel di sekitarnya. Ini mengurangi detail halus dan noise dalam gambar.
*  Mengurangi Noise: Kernel rata-rata efektif dalam mengurangi noise acak, tetapi dapat mengaburkan tepi dan detail penting.
*  Efek Blur: Gambar akan terlihat lebih buram dibandingkan dengan gambar asli, yang merupakan hasil dari penggabungan informasi dari piksel-piksel tetangga.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Membaca citra grayscale
image_path = '/content/zam.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# 2. Melakukan Transformasi Fourier
f = np.fft.fft2(image)
fshift = np.fft.fftshift(f)
magnitude_spectrum = np.log(np.abs(fshift) + 1)

# 3. Menampilkan magnitude spectrum
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Magnitude Spectrum')
plt.imshow(magnitude_spectrum, cmap='gray')
plt.axis('off')

# 4. Melakukan invers Transformasi Fourier
f_ishift = np.fft.ifftshift(fshift)
image_back = np.fft.ifft2(f_ishift)
image_back = np.abs(image_back)

# 5. Menampilkan citra hasil invers
plt.subplot(1, 2, 2)
plt.title('Citra Hasil Invers Transformasi Fourier')
plt.imshow(image_back, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

Penjelasan Kode :

*   Membaca Citra: Menggunakan cv2.imread() untuk membaca citra dalam mode grayscale.
*   Transformasi Fourier: np.fft.fft2() digunakan untuk melakukan Transformasi Fourier dua dimensi pada citra. Kemudian, np.fft.fftshift() digunakan untuk memindahkan komponen frekuensi nol ke tengah spektrum.
*   Magnitude Spectrum: Magnitude spectrum dihitung dengan mengambil logaritma dari nilai absolut dari hasil transformasi, yang memberikan gambaran tentang kekuatan frekuensi yang ada dalam citra.
*   Invers Transformasi Fourier: np.fft.ifft2() digunakan untuk mengembalikan citra dari domain frekuensi ke domain spasial. Nilai absolut diambil untuk mendapatkan citra yang dapat ditampilkan.
*   Menampilkan Hasil: Menggunakan Matplotlib untuk menampilkan magnitude spectrum dan citra hasil invers.

Informasi dari Magnitude Spectrum
Magnitude spectrum memberikan informasi penting mengenai frekuensi yang terkandung dalam citra:
*   Frekuensi Rendah: Terletak di tengah magnitude spectrum, frekuensi rendah biasanya berkaitan dengan informasi global dan struktur besar dalam citra.
*  Frekuensi Tinggi: Terletak di tepi magnitude spectrum, frekuensi tinggi berkaitan dengan detail halus, tepi, dan noise dalam citra.
*    Analisis Citra: Dengan menganalisis magnitude spectrum, kita dapat memahami komponen frekuensi yang mendominasi citra, yang berguna dalam berbagai aplikasi seperti pengenalan pola, kompresi citra, dan penghilangan noise.











```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Membaca citra grayscale
image_path = '/content/zam.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# 2. Menambahkan konstanta untuk pencerahan
constant = 50  # Nilai konstanta untuk pencerahan
brightened_image = image + constant

# 3. Melakukan clipping untuk memastikan nilai piksel tidak melebihi 255
brightened_image = np.clip(brightened_image, 0, 255).astype(np.uint8)

# 4. Menampilkan citra asli dan citra yang telah diperoleh setelah pencerahan
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Citra Asli')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Citra Setelah Pencerahan')
plt.imshow(brightened_image, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

Penjelasan Kode :
*   Membaca Citra: Menggunakan cv2.imread() untuk membaca citra dalam mode grayscale.
*   Menambahkan Konstanta: Kita menambahkan nilai konstanta (dalam contoh ini, 50) ke setiap piksel dalam citra.
*   Clipping: Menggunakan np.clip() untuk memastikan bahwa semua nilai piksel berada dalam rentang 0 hingga 255. Jika ada nilai yang lebih dari 255, nilai tersebut akan dipotong menjadi 255.
*   Menampilkan Hasil: Menggunakan Matplotlib untuk menampilkan citra asli dan citra yang telah diperoleh setelah pencerahan.

Perubahan yang Terjadi Setelah Pencerahan
Setelah melakukan pencerahan pada citra :
*   Kecerahan Meningkat: Citra akan terlihat lebih terang karena setiap piksel telah ditambahkan dengan nilai konstanta. Ini membuat detail dalam citra lebih terlihat, terutama pada area yang sebelumnya gelap.
*  Detail yang Hilang: Jika nilai konstanta terlalu tinggi, beberapa detail di area terang mungkin hilang karena clipping. Misalnya, jika piksel sudah dekat dengan 255, penambahan nilai akan membuatnya tetap 255, sehingga detail tersebut tidak dapat ditangkap.
*   Kontras Berubah: Pencerahan dapat mengubah kontras citra, tergantung pada distribusi nilai piksel awal. Area gelap akan menjadi lebih terang, tetapi area terang mungkin kehilangan beberapa nuansa.
